```python
import numpy as np
import tensorflow as tf
import cv2

# Load your TensorFlow Lite model
model_path = r"C:\Users\aliha\Desktop\FYP\Tumor_detection\tumor_detection_model.tflite"  # Replace with the actual full path
interpreter = tf.lite.Interpreter(model_path=model_path)
interpreter.allocate_tensors()

# Get input and output details
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# Load and preprocess the input image using OpenCV
#image_path = r"C:\Users\aliha\Desktop\FYP\Project\test\benign\melanoma_9691.jpg"
image_path =  r"C:\Users\aliha\Desktop\FYP\Project\test\malignant\melanoma_10131.jpg"
image = cv2.imread(image_path)
image_resized = cv2.resize(image, (224, 224))  # Resize to match model input shape
input_image = image_resized.astype(np.float32) / 255.0  # Convert to FLOAT32 and normalize

# Add batch dimension
input_image = np.expand_dims(input_image, axis=0)

# Set the input tensor
interpreter.set_tensor(input_details[0]['index'], input_image)

# Run inference
interpreter.invoke()

# Get the output tensor
output = interpreter.get_tensor(output_details[0]['index'])

# Post-process the output
prediction = np.argmax(output)
confidence = output[0][prediction]

# Define a threshold for confidence (you can adjust this)
threshold = 0.1

# Make the prediction based on the threshold
result = "Malignant" if confidence > threshold else "Benign"

# If the tumor is malignant, create a bounding box at the center of the image
if result == "Malignant":
    # Get the dimensions of the image
    height, width, _ = image_resized.shape

    # Calculate the coordinates for the bounding box at the center
    box_size = 100  # You can adjust the size of the bounding box
    x_center = (width - box_size) // 2
    y_center = (height - box_size) // 2

    # Draw the bounding box on the original image
    cv2.rectangle(image_resized, (x_center, y_center), (x_center + box_size, y_center + box_size), (0, 255, 0), 2)

    # Extract the coordinates of the bounding box
    print(f"Bounding box coordinates: ({x_center}, {y_center}), ({x_center + box_size}, {y_center + box_size})")

    # Display the image with bounding box
    cv2.imshow("Tumor Detection", image_resized)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

else:
    # If the tumor is benign, just print the result
    print(f"The tumor is classified as: {result} (Confidence: {confidence * 100:.2f}%)")
```
